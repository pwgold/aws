```python
def plot_summary(rates, do_pca=True, use_corr=False, input_title=''):
    if type(rates) is pd.Series:
        do_pca=False
        rates = pd.DataFrame(rates)
    # Input Params
    V = None
    if do_pca:
        # PCA on LEVELS or Changes??
        if use_corr:
            #C = rates.diff().corr()
            C = rates.corr()
        else:
            #C = rates.diff().cov()
            C = rates.cov()
        V,_ = pcacov(C)            # second_argument is sqrt variances
        pca_rates = rates.dot(V.T) #.ffill(limit=10).dropna() # shouldn't be any nans in rates or pcacov won't work
        #V.plot(kind='bar')
        figure()
        V.plot(kind='bar', title='Principal Component Weights')
    else:
        pca_rates = rates
    
    pca_returns = pca_rates.diff().dropna()
    # Fit OU and print results
    output = fit_ou(pca_returns)
    #output['scores'].plot()
    #print np.round(output['stats'],3)
    # Plot OU for each PC_IDX
    output_axes = []
    for pca_idx in pca_returns.columns:
        ax = plot_ou(pca_returns[pca_idx], output['stats'][pca_idx], input_title=input_title)
        output_axes.append(ax)
    table = np.round(output['stats'],3)
    table.index.name = 'Principal Component'
    b = pca_returns.cumsum().tail(1)
    b.index = ['%s\nvalue'%pd.to_datetime(e).strftime('%Y-%m-%d') for e in b.index.values] # convert to string
    b = np.round(b,2)
    table = pd.concat([table.T, b.T], axis=1)
    return table, V, pca_returns, output_axes


###################
# Estimate OU parameters of A_t+1 = a + b*A_t + ζ_t+1, where A_t = Sum(e_t), and
# e_t is s.t. r_t = β0 + β*x_t + e_t (t=1,...,T=len(r)).
# Set α = β0/∆t = β0 ∗ 252. (The annualized sample mean of the residual returns)
# Then a = m*(1-e^-κ*∆t), b = e^-2κ*∆t, and var(ζ) = σ^2 * (1-e^-2κ*∆t)/2κ, and so
# kappa = -log(b)/tau                           # speed of mean-reversion 
# m = a / (1-b)                                 # equilibrium mean
# sigma_eq = np.sqrt(var_tau / (1 - b**2))      # equilibrium sqrt variance
# sigma = np.sqrt(2*kappa)*sigma_eq             # ??
# reversion_days = 1/(kappa*tau)                # log(e) / -log(b)    
# halflife = np.log(0.5)/(-kappa*tau)           # -log(0.5) / -log(b) 
# So halflife / reversion_days = -log(0.5)/log(e) = log(2), and
# kappa = (1/reversion_days)*(252 days/year) ~ cycles per year 
# Very optimistically, I guess total expected return is
# capacity = sigma_eq * (1.0/reversion_days * 1.0/tau) # vol_per_opportunity * opportunities_per_year
# Set Score: score = (A - m)/sigma_eq
# adf_pvalue = adf_test(A)
#
# References:
# https://www.math.nyu.edu/faculty/avellane/AvellanedaLeeStatArb071108.pdf
# https://www.math.nyu.edu/faculty/avellane/Lecture6Quant.pdf
###################

def pcacov(F):
    from scipy.linalg import svd, cholesky
    n,k = F.shape
    assert n == k
    _,s,V = svd(cholesky(F)) # e.g., if F = r.cov(), s = r.dot(V.T).std()
    V = pd.DataFrame(V, index=range(k), columns=F.columns)
    s = pd.Series(s, index=range(k)) # std
    return V,s


def __fit_ou(e, tau=1/252.0, x=None, return_all=False):
    ###################
    # Estimate OU parameters of A_t+1 = a + b*A_t + ζ_t+1, where A_t = Sum(e_t), and
    # e_t is s.t. r_t = β0 + β*x_t + e_t (t=1,...,T=len(r)).
    # If X is not None, assume input e is r_t; then X should be risk factor exposures or common factor returns.
    # TODO: o Fit r with dim > 1 (??)
    #       o include drift parameter
    
    # Squeeze r incase it's a 1D dataframe
    e = e.squeeze()     
    if x is not None: 
        # So e is r and r,x are s.t. r = β0 + β*x + e
        reg = pd.ols(y=e, x=x)
        e = reg.resid                      # NB: overwriting input e
        alpha = reg.beta['intercept']/tau  # Set α = β0/∆t = β0 ∗ 252. (The annualized sample mean of the residual returns)
        exposures = reg.beta
    else:
        alpha = e.mean()                   # Ibid.
        exposures = pd.Series(np.nan)      # NB: we don't know β
        
    # Now define auxiliary OU process, A_t = Sum_j^k e_j k=1,...,T2 (NB: if T=T2, X_T=0)
    A = e.cumsum()
    # And estimate, A_t+1 = a + b*A_t + ζ_t+1
    reg = pd.ols(y=A, x=A.shift(1))
    a = reg.beta.values[1]
    b = reg.beta.values[0]
    if b < 0: 
        print 'beta is less than 0, returning None'
        return None
    var_tau = reg.resid.var() 
    # Then a = m*(1-e^-κ*∆t), b = e^-2κ*∆t, and var(ζ) = σ^2 * (1-e^-2κ*∆t)/2κ, and so
    kappa = -np.log(b)/tau                           # speed of mean-reversion 
    m = a / (1-b)                                 # equilibrium mean
    sigma_eq = np.sqrt(var_tau / (1 - b**2))      # equilibrium sqrt variance
    sigma = np.sqrt(2*kappa)*sigma_eq             # ??
    reversion_days = 1/(kappa*tau)                # log(e) / -log(b)    
    halflife = np.log(0.5)/(-kappa*tau)           # -log(0.5) / -log(b) 
    # So halflife / reversion_days = -log(0.5)/log(e) = log(2), and
    # kappa = (1/reversion_days)*(252 days/year) ~ cycles per year 
    # Very optimistically, I guess total expected return is
    capacity = sigma_eq * (1.0/reversion_days * 1.0/tau) # vol_per_opportunity * opportunities_per_year

    # Set Score
    score = (A - m)/sigma_eq
    
    adf_pvalue = adf_test(A)
    
    # Setup output
    stats = pd.Series({'m':m, 'kappa':kappa, 
                       'sigma':sigma, 'sigma_eq':sigma_eq,
                       'reversion_days':reversion_days, 'halflife':halflife,
                       'alpha':alpha, 'a':a, 'b':b,
                       'capacity':capacity,
                       'adf_pvalue':adf_pvalue})
                       # 'tau':tau}) # NB: Might as well carry tau around ?
    
    # NB: assert np.allclose(((Y.cumsum() - stats['m']) / stats['sigma_eq']), scores)
    
    if return_all:
        return stats, reg, exposures, score
    return stats


def fit_ou(Y, tau=1/252.0, return_all=True):
    # Just a wrapper for __fit_ou, but it doesn't handle X, so exposures here will always be np.nan
    if type(Y) is pd.DataFrame:
        output = {'stats':{}, 'scores':{}} # , 'exposures':{}}
        for col in Y.columns:
            o = fit_ou(Y[col], tau=tau, return_all=True) # NB: request everything
            for key in output.keys():
                output[key][col] = o[key]
        for key in output.keys():
            output[key] = pd.DataFrame(output[key])
        if not(return_all):
            return output['scores']
        else:
            return output
    else:
        assert type(Y) is pd.Series, 'Input Y should be a pandas Series or DataFrame'
        # Assuming Y is a pd.Series
        stats,_,exposures,scores = __fit_ou(Y, tau=tau, return_all=True)
        if not(return_all):
            return scores
        else:
            output = {'scores':scores,
                      'stats':stats}#,
    #                  'exposures':exposures}
            return output


def plot_ou(returns, stats=None, tau=1/252.0, input_title=''):
    # NB: returns is I(0), and we want cumulative_returns that are I(1)
    if stats is None:
        stats = fit_ou(returns, tau)['stats']

    y = pd.DataFrame(returns.cumsum().copy())           # NB: Make copy not modify cumulative_returns (??)
    #y = pd.DataFrame((1+returns).cumprod().copy()-1)   # NB: Make copy not modify cumulative_returns (??)
    y['m'] = stats['m']
    y['m + sigma_eq'] = y['m'] + stats['sigma_eq']
    y['m - sigma_eq'] = y['m'] - stats['sigma_eq']
    y['m + 2*sigma_eq'] = y['m'] + 2*stats['sigma_eq']
    y['m - 2*sigma_eq'] = y['m'] - 2*stats['sigma_eq']
    ttl = '%s Plot of Cumulative Returns (Reversion Days: %.3f, Kappa: %.3f, Halflife: %.3f)'%(input_title, 
                                                                                               stats['reversion_days'],
                                                                                               stats['kappa'],
                                                                                               stats['halflife'])
    ax = y.plot(style=['b','k','r--','r--', 'g--', 'g--'], title=ttl)
    return ax
    # plt.legend({y.columns[0], 'sigma_eq', 'm'})

    
    
def plot_pca(rates, use_corr=False, plot_cumulative=False, take_cov=True):
    if take_cov:
        if use_corr:
            F = rates.corr()
        else:
            F = rates.cov()
    else:
        F = rates # input was cov or corr matrix
    V,s = pcacov(F)

    figsize(20,5)
    V.plot(kind='bar')
    xlabel('PC')
    ylabel('Weight')

    figure()
    s.plot(kind='bar')
    ylabel('standard deviation')
    ((s**2).cumsum()/(s**2).sum()).plot(secondary_y=True, style='b--')
    ylabel('cumulative % variance explained')
    figure()
    pcs = pd.DataFrame(rates.dot(V.T), columns=V.index, index=rates.index)
    if plot_cumulative:
        pcs.cumsum().plot() # NB converts from pd.DataFrame to np.array / matrix
    else:
        pcs.plot(secondary_y=[0,1])
    return pcs
        
import statsmodels.tsa.stattools as stt
def adf_test(Y, regression='c', rejection_level=None):
    '''
    Return pvalue of adf test for stationarity, or boolean if pvalue < rejection_level
    Inputs:
     regression : str {'c','ct','ctt','nc'}
        Constant and trend order to include in regression
        * 'c' : constant only
        * 'ct' : constant and trend
        * 'ctt' : constant, and linear and quadratic trend
        * 'nc' : no constant, no trend
    '''
    # Ensure Y is series without nans
    Y = Y.dropna().squeeze()  
    assert type(Y) is pd.Series
    adf = stt.adfuller(Y, regression=regression) # regresults=True)[-1].resols.summary()
    pvalue = adf[1]
    if rejection_level is not None:
        return pvalue < rejection_level
    else:
        return pvalue
```

```python
def do_pca(a):
    # a = pd.DataFrame(np.random.randn(100, 6)) 
    #a = us_fi_dataD.diff().dropna() # .diff()['2015']
    # a = a / a.std()
    # F = a.corr() is the same as a0.cov() for a0 = a / a.std() # normalized
    F = a.cov()

    L = np.linalg.cholesky(F)
    assert np.allclose(F, L.dot(L.T))
    U,s,V = np.linalg.svd(L)
    S = np.zeros(L.shape) 
    S[:L.shape[1], :L.shape[1]] = np.diag(s)
    assert np.allclose(L, np.dot(U, np.dot(S, V)))
    loadings = pd.DataFrame(U, index=a.columns, columns=range(a.shape[1])).T # NB: Transpose
    pcs = a.dot(loadings.T)
    assert np.allclose(pcs, a.dot(U)) # unnecssary: 
    assert np.allclose(a.dot(U)[0] - a.dot(U.T[[0]].T),0) # columns of U are PC loadings
    assert np.allclose(F, np.dot(U,np.dot(S,V)).dot(np.dot(U,np.dot(S,V)).T)) # reconstruct
    assert np.allclose(pcs.std(), s)                    # U are loadings, a.dot(U) are PCs
    assert np.allclose(pcs.corr(), np.eye(a.shape[1]))
    w = (a - a.mean()).dot(U)/s
    assert np.allclose(w.mean(), 0)
    assert np.allclose(w.cov(), np.eye(a.shape[1]))
    assert np.allclose(w.dot(L.T).cov(), F)
    return loadings, pcs

def plot_pca2(a):
    #figsize(20,4)
    #figure()
    #a.cumsum().plot()
    loadings, pcs = do_pca(a)
    
    figure()
    loadings.plot(kind='bar', title='PC Loadings')
    xlabel('PC')
    ylabel('Weight')

    s = pcs.std()
    figure()
    ((s**2).cumsum()/(s**2).sum()).plot(secondary_y=True, style='b--', title='%% of Variance Explained')
    ylabel('PC Standard Deviation / Variance')
    s.plot(kind='bar')
    ylabel('standard deviation')
    figure()
    pcs.cumsum().plot(title='Cumulative PC "returns"', secondary_y=0) # NB converts from pd.DataFrame to np.array / matrix
    return loadings, pcs
```
