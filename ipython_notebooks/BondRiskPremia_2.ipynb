#### Helper Functions

```python
def pcacov(F):
    from scipy.linalg import svd, cholesky
    n,k = F.shape
    assert n == k
    _,s,V = svd(cholesky(F)) # e.g., if F = r.cov(), s = r.dot(V.T).std()
    V = pd.DataFrame(V, index=range(k), columns=F.columns)
    s = pd.Series(s, index=range(k))
    return V,s
```

```python
def plot_pca(rates, use_corr=False, plot_cumulative=False):
    if use_corr:
        F = rates.corr()
    else:
        F = rates.cov()
    V,s = pcacov(F)

    V.plot(kind='bar')
    xlabel('PC')
    ylabel('Weight')

    figure()
    np.sqrt(s).plot(kind='bar')
    ylabel('standard deviation')
    (s.cumsum()/s.sum()).plot(secondary_y=True, style='b--')
    ylabel('cumulative % variance explained')
    figure()
    if plot_cumulative:
        plot(rates.dot(V.T).cumsum()) # NB converts from pd.DataFrame to np.array / matrix
    else:
        # plot(rates.dot(V.T))
        pd.DataFrame(rates.dot(V.T), columns=V.index, index=rates.index).plot(secondary_y=0)
```

#### CP Factor Not Subsumed by Level, Slope, Curvature

```python
rotation, std_dev = pcacov(forward_rates.cov())    # use full and not annual forward rates
pca_forward_rates = forward_rates.dot(rotation.T)  # ['level', 'slope', 'curve', 'factor4', 'factor5']
```

```python
# Table 2 on page 19 of Decomposing the Yield Curve
# NB: My factor5 may be their factor4. Also they have a much higher r2 for returns ~ X, 
# which helps motivate their point more about being required to include the return forecasting factor X.
def calculate_table(pca_forward_rates, Y=None):
    # Regression of Y (average excess) returns on return forecasting factor
    # If Y is None, use average annual returns.
    # Use Y = return_forecasting_factor to check ability to "explain" return forecasting factor
    # or Y = returns
    if Y is None:
        Y = annual_excess_returns.mean(axis=1)   # explain average annual returns
    reg = pd.ols(y=Y, # annual_excess_returns.mean(axis=1),
                 x=return_forecasting_factor)
    print '\nReturn forecasting factor:'
    print reg.t_stat
    print np.round(pd.Series({'beta':reg.t_stat['x'],
                              'r2':reg.r2}),3)

    for include_return_forecasting_factor in [False, True]:
        betas = {}
        tstats = {}
        r2 = {}
        for idx in pca_forward_rates.columns:
            X = pca_forward_rates.ix[:, :idx]        # or just ,idx]) for effect of individual pc
            if include_return_forecasting_factor:
                X['X'] = return_forecasting_factor   # include return forecasting factor along with level,slope,etc.
            
            if Y is None:
                Y = annual_excess_returns.mean(axis=1)   # explain average annual returns
                
            reg = pd.ols(y=Y,
                         x=X)
            betas[idx] = reg.beta
            tstats[idx] = reg.t_stat
            r2[idx] = reg.r2
        betas = pd.DataFrame(betas).T
        betas['r2'] = pd.Series(r2)
        # betas.ix['% var(f)'] = std_dev/std_dev.sum()
        betas = np.round(betas,3).replace(np.nan,'')
        tstats = np.round(pd.DataFrame(tstats),3).replace(np.nan,'').T

        print '\nPCA betas:'
        print betas

        print 'PCA tstats:'
        print tstats

    df = pca_forward_rates.ix[annual_indices, [2,3]]
    df['X'] = return_forecasting_factor
    df.plot(secondary_y='X', style=['b','r','k--'])
```

```python
plot_pca(forward_rates)
```

```python
calculate_table(pca_forward_rates) # Y=annual_excess_returns.mean(axis=1))
```

```python
# orthogonalize forward rates with respect to return forecasting factor
e = {} # resid
c = {} # intercept
for col in forward_rates.columns:
    reg = pd.ols(y=forward_rates[col], x=return_forecasting_factor)
    e[col] = reg.resid
    c[col] = reg.beta['intercept']
c = pd.Series(c)    # NB: Seems like in CP they have a single c. Panel regression? (Pooled OLS didn't work)
e = pd.DataFrame(e) # e itself is not now invertible. In CP they use more forward rates; that may expalin it. 

e_rotation,e_stds = pcacov(e.ix[:,:4].cov())      # e itself is not invertible ... in CP they use more forward rates...
pca_ortho_forwards = e.ix[:,:4].dot(e_rotation.T) + c.ix[:4].dot(e_rotation.T) 
```
