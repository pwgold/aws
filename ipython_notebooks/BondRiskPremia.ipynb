#### Helper Functions

```python
def get_maturity(yields):
    if type(yields) is pd.DataFrame:
        maturity = [float(yld.rstrip('Y')) for yld in yields.columns]
        maturity_df = pd.DataFrame(data=[maturity for idx in range(len(yields.index))],
                                   index = yields.index,
                                   columns = yields.columns)
    elif type(yields) is pd.Series:
        maturity = [float(yld.rstrip('Y')) for yld in yields.index]
        maturity_df = pd.Series(data=maturity, index=yields.index)
    return maturity_df

def yields_to_prices(yields, maturity_df=None):
    # price = sum( CF_j / Y**j ) for all j
    # so for discount bonds
    # prices = 1 / Y**j
    if maturity_df is None:
        maturity_df = get_maturity(yields)
    prices = 1 / ((1+yields) ** maturity_df)
    return prices

def prices_to_yields(prices, maturity_df=None):
    if maturity_df is None:
        maturity_df = get_maturity(prices)              # In years
    yields = (1.0 / prices)**(1.0/maturity_df) - 1
    return yields
```

#### Calculate Prices, Forward Rates, and Returns

```python
####
# Calculate prices from these rates
####
prices = yields_to_prices(yields)
assert np.allclose(yields.replace(np.nan, 0), prices_to_yields(prices).replace(np.nan,0)) # just check function works

log_prices = log(prices).ffill(limit=2)
assert np.allclose(log_prices.apply(np.isnan).sum(), 0) # no nans
assert np.allclose((-1 / get_maturity(log_prices)) * log_prices, log(1+yields)) # sanity check
```

```python
def shift_horizontally(input_df):
    df = pd.DataFrame(input_df[input_df.columns[0:-2]].values,
                     index=input_df.index, 
                     columns=input_df.columns[1:-1])
    df[input_df.columns[0]] = np.nan
    return df
```

```python
# Forward rates as they're indexed in CP:
# f^n is n-1 --> n year (log) forward rate
original_calc = pd.DataFrame({'1Y':0 - log_prices['1Y'],
                              '2Y':log_prices['1Y'] - log_prices['2Y'],
                              '3Y':log_prices['2Y'] - log_prices['3Y'],
                              '4Y':log_prices['3Y'] - log_prices['4Y'],
                              '5Y':log_prices['4Y'] - log_prices['5Y']})

#forward_rates = log_prices.shift(1, axis=1).replace(np.nan, 0) - log_prices
forward_rates = shift_horizontally(log_prices).replace(np.nan, 0) - log_prices
forward_rates = forward_rates[['1Y','2Y','3Y','4Y','5Y']]
assert np.allclose(forward_rates, original_calc)

# Generate rolling "annual" returns
# r_t+1 = p_t+1^(n-1) - p_t^(n) is t --> t+1 return
original_calc = pd.DataFrame({'2Y':log_prices['1Y'].shift(-252) - log_prices['2Y'],
                              '3Y':log_prices['2Y'].shift(-252) - log_prices['3Y'],
                              '4Y':log_prices['3Y'].shift(-252) - log_prices['4Y'],
                              '5Y':log_prices['4Y'].shift(-252) - log_prices['5Y']})

#rolling_annual_returns = log_prices.shift(1, axis=1).replace(np.nan, 0).shift(-252) - log_prices
rolling_annual_returns = shift_horizontally(log_prices).replace(np.nan, 0).shift(-252) - log_prices
rolling_annual_returns = rolling_annual_returns[['2Y','3Y','4Y','5Y']]
assert np.allclose(rolling_annual_returns.dropna(), original_calc.dropna()) # There are NaN at the end

# Subtract the 1Y rate to get excess returns
rolling_annual_excess_returns = rolling_annual_returns.subtract(yields['1Y'], axis=0)
```

#### Analysis

##### Is there a risk premium on average

```python
# Use 1964 - 2102 to match table on p. 533 of notes
# Note: o Get yearly indices to use non-overlapping data
#       o Because we had no 4Y yields, this is messing up 4Y,5Y values in first table and 5Y values in second table
annual_indices = sorted(list(set([yields.ix['%s'%year:].index[0] for year in range(1964,2013)])))
annual_log_prices = log_prices.ix[annual_indices,:]

# calculate yearly returns and excess returns
# NB: clobbering r variable above
# TODO: o Interpret this '1Y'
annual_returns = pd.DataFrame({#'1Y':0 - annual_log_prices['1Y'], 
                               '2Y':annual_log_prices['1Y'].shift(-1) - annual_log_prices['2Y'],
                               '3Y':annual_log_prices['2Y'].shift(-1) - annual_log_prices['3Y'],
                               '4Y':annual_log_prices['3Y'].shift(-1) - annual_log_prices['4Y'],
                               '5Y':annual_log_prices['4Y'].shift(-1) - annual_log_prices['5Y']})  

annual_excess_returns = annual_returns.subtract(yields['1Y'],axis=0).ix[annual_indices,:]

df = pd.DataFrame({'E[y^n]':yields.ix[annual_indices,:].mean()*100,
                   'E[y^n - y^1]':yields.subtract(yields['1Y'], axis=0).ix[annual_indices,:].mean()*100,
                   'E[r^n - y^1]':annual_excess_returns.mean()*100,
                   't statistic': sqrt(annual_excess_returns.shape[0])*annual_excess_returns.mean()/annual_excess_returns.std(),
                   'sigma[r^n - y^1]' : annual_excess_returns.std()*100,
                   '"Sharpe"' : annual_excess_returns.mean() / annual_excess_returns.std()}).T

print 'Is there a risk premium on average?'
df = np.round(df,4).replace(np.nan, '')
print df.ix[['E[y^n]', 'E[y^n - y^1]',  'E[r^n - y^1]', 't statistic', 'sigma[r^n - y^1]', '"Sharpe"'],
            ['1Y','2Y','3Y','4Y','5Y']] #,'7Y','10Y']]
```

##### Expectations Failures 

```python
print 'Expectations failures (Fama Bliss)'
print 'Forecasting one year returns on n-year bonds'
annual_forward_spot_spreads = forward_rates.subtract(yields['1Y'], axis=0).ix[annual_excess_returns.index]

table_df = {}
for mat in annual_excess_returns.columns:
    reg = pd.ols(y=annual_excess_returns[mat], 
                 x=annual_forward_spot_spreads[mat])
    table_df[mat] = pd.Series({'b':reg.beta['x'],
                               'sigma(b)':reg.std_err['x'],
                               'R^2':reg.r2})
df = np.round(pd.DataFrame(table_df).T, 3)
print '\nTable 1'
print 'annual_excess_returns[mat] ~ annual_forward_spot_spreads[mat]'
print df[['b','sigma(b)','R^2']]
```

```python
# Forecasting one year rates n years from now
# y_t+n-1^(1) - y_t^(1) = a + b*(f_t^(n) - y_t^(1)) + e_t+1
# NB: 4Y rate was purely interpolated 
table_df = {}
for mat in annual_excess_returns.columns:
    n = int(mat.replace('Y',''))
    reg = pd.ols(y = yields.ix[annual_indices, '1Y'].shift(-(n-1)) - yields.ix[annual_indices,'1Y'],
        x = annual_forward_spot_spreads[mat])
    table_df[mat] = pd.Series({'b':reg.beta['x'],
                               'sigma(b)':reg.std_err['x'],
                               'R^2':reg.r2})
df = np.round(pd.DataFrame(table_df).T, 3)
print '\nTable 2'
print 
print df[['b','sigma(b)','R^2']]
```

```python
print '''
# Notes from CP:
# In the Table 1, b "should be" 0 and in the second table b "should be 1". 
# Instead, in Table 1, b = 1: Expected excess returns move 1 : 1 with the forward - yield spread.
# Long bonds don't earn more *on average* (very first table). But there are times when they are expected to earn more or less.
# In Table 2, b = 0: a forward rate 1% higher than the spot rate should mean spot rate rises. Instead, it means that the 2 year # bond earns 1% more over the next year on average. (Note: 0.891 + 0.109 = 1 mechanically. (p.534).)
# So an "expected offsetting adjustment doesn't happen". Therefore, for bonds, 
# Is forward - yield high? Then Yield should rise. It doesn't (at least for a few years). So "buy yield".
# (An upward sloping yield curve results in rising rates, but not soon enough - you earn the yield in the meanime.)
# So "prices reflect big risk premiums, not expected changes in rates".
'''
```

#### CP - Tent Shaped Factor

```python
# Unrestricted Regression
############### 
# Rather than using individual forward - spot (1Y) spreads as in FB above, 
# regress annual returns on *all* forward rates.
# NB: Can't exactly recreate the unrestricted chart on p. 540. But we had no 4Y yield and was missing ~2000 days of 2Y yield. 
# So inlcuding just the 1Y,3Y,5Y tenors we see the shape:
regression_tenors = ['1Y','3Y','5Y']
#regression_tenors = ['1Y','2Y','3Y','4Y','5Y']

regression_results = {}
for tenor in annual_excess_returns:
    reg = pd.ols(y=annual_excess_returns[tenor], 
                 x=forward_rates.ix[annual_indices, regression_tenors])
    regression_results[tenor] = reg

betas_df = pd.DataFrame(dict((tenor, regression_results[tenor].beta[regression_tenors]) for tenor in annual_excess_returns.columns))
regression_r2 = pd.Series(dict((tenor, regression_results[tenor].r2) for tenor in annual_excess_returns.columns))
regression_r2.name = 'R-squared'
```

```python
betas_df.plot()
print '################'
print regression_r2
print '################'
```

```python
# Restricted Regression
##########
# "A single factor for expected bond returns"
# Regress average annual excess returns against forward rates (to uncover tent shaped factor betas), 
# and then regress individual annual excess returns against (restricted) tent shaped factor.
# NB: p. 542, they include intercept in gamma, and presumably in regression as well, but again,
#     not really matching the intercept here
############
regression_tenors = ['1Y','3Y','5Y'] # just a reminder
#regression_tenors = ['1Y','2Y','3Y','4Y','5Y']

reg = pd.ols(y = annual_excess_returns.mean(axis=1),
             x = annual_forward_rates[regression_tenors],
             intercept=True)
gamma = reg.beta # tent shaped factor, with intercept

# Restricted regression
F = annual_forward_rates[regression_tenors].copy()
F['intercept'] = 1
return_forecasting_factor = F.dot(gamma)

restricted_regression_results = {}
for tenor in annual_excess_returns.columns:
    reg = pd.ols(y = annual_excess_returns[tenor],
                 x = return_forecasting_factor,
                 intercept=True)
    restricted_regression_results[tenor] = reg
    restricted_regression_results[tenor] = pd.Series({'b_n':reg.beta['x'],
                                                      'R^2':reg.r2})

df = pd.DataFrame(restricted_regression_results).T[['b_n', 'R^2']]  # b_n is the scaling factor we apply to the single factor
df['R^2 (unrestricted)'] = regression_r2                            # unrestricted regression r2's from above
print df
```

```python
####### 
# Plot Restricted Betas
# NB: Shape of originals is mostly preserved, but maybe less than reported by CP 
#######
restricted_betas_df = pd.DataFrame(dict((tenor, betas_df[tenor]*restricted_regression_results[tenor]['b_n'])
                                        for tenor in annual_excess_returns.columns))
restricted_betas_df.plot(title='restricted regression betas')
```
