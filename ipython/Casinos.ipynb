```python
# load files
files = glob.glob('aws/collect/InteractiveBrokers/data.CLEAN/data.usa/*.usa.txt')
file_dict = dict((e.split('/')[-1].split('%')[0], e) for e in files)
```

```python
def get_ticker_borrow(ticker):
    output = {}
    for date, filename in file_dict.items():
        try:
            output[date] = pd.read_csv(filename, sep='|', skiprows=0, header=True, index_col=0).ix[ticker]
        except Exception as e:
            # print date, e
            ""
    
    df = pd.DataFrame(output).T
    df.AVAILABLE = df.AVAILABLE.apply(lambda x : float(x.replace('>','')))
    return df
```

```python
short_data = {}
for ticker in ['BYD','PNK','SPY']:
    short_data[ticker] = get_ticker_borrow(ticker)
```

```python
borrow_rate = pd.DataFrame(dict((tk, short_data[tk]['REBATERATE']) for tk in short_data.keys()))
```

```python
borrow_rate.plot(secondary_y='SPY')
```
