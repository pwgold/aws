```python
import calendar
cal = pd.date_range(#adj_close.dropna().index[0],
                    pd.datetime(1989,12,1), 
                    #pd.datetime(2015,5,1),
                    pd.datetime.today(),
                    freq='B')
```

```python
def is_month_end(dt, next_dt):
    # cal[0].day == calendar.monthrange(cal[0].year, cal[0].month)[1]
    last_day_of_month = calendar.monthrange(dt.year, dt.month)[1]
    if dt == last_day_of_month:
        return True
    if next_dt > pd.datetime(dt.year, dt.month, last_day_of_month):
        return True
    return False
        
```

```python
#month_end = pd.Series(data=[is_month_end(e) for e in cal],
month_end = pd.Series(data=[is_month_end(cal[idx],cal[idx+1]) for idx in range(len(cal)-1)],
                      index=cal[range(len(cal)-1)])

reversed_month_end = month_end.sort_index(ascending=False)
```

```python
def days_till_month_end(dt):
    if dt not in month_end.index:
        return np.nan
    
    L = list(month_end.ix[dt:])
    if True in L:
        return L.index(True) # following day is month end, so we're at T-1
    else:
        return np.nan
```

```python
def days_since_month_end(dt):
    if dt not in reversed_month_end.index:
        return np.nan
    
    L = list(reversed_month_end.ix[dt:])
    if True in L:
        return L.index(True) # prior month end was 21 business days ago, so we're at T+21
    else:
        return np.nan
```

```python
# General idea is that cash distributions are made on T (last business day of the month), so
# "average returns are low from T-8 to T-4 (selling pressure to meet month-end cash distributions), 
# high from T-3 to T-1 (return reversal), 
# high from T to T+3 (buying pressure from automated reinvestment), 
# low from T+4 to T+8 (return reversal).
```

```python
# TODO: x sanity check that not more than one of these conditions are never met
#       o combine these in single "days since month end" with negative values included
def period_0(s):
    return 4 <= s['days_till_month_end'] <= 8 # pension funds sell until T-4 so they can make cash distributions on T

def period_1(s):
    return 1 <= s['days_till_month_end'] <= 3

def period_2(s):
    return 0 <= s['days_since_month_end'] <= 3  # earns return from buying at the close on day *before* last business day of the month
    
def period_3(s):
    return 4 <= s['days_since_month_end'] <= 8

def get_periods(s):
    #if type(s) is not pd.tslib.Timestamp:
    #    if np.isnan(s['days_since_month_end']) or np.isnan(s['days_till_month_end']):
    #        return np.nan
    if period_0(s):
        assert not(any([period_1(s), period_2(s), period_3(s)]))
        return 0
    elif period_1(s):
        assert not(any([period_0(s), period_2(s), period_3(s)]))
        return 1
    elif period_2(s):
        assert not(any([period_0(s), period_1(s), period_3(s)]))
        return 2
    elif period_3(s):
        assert not(any([period_0(s), period_1(s), period_2(s)]))
        return 3
    else:
        return 4
```

```python
periods_df = pd.DataFrame(data = adj_close.index,
                          index = adj_close.index,
                          columns = ['date'])
   
periods_df['days_since_month_end'] = periods_df['date'].apply(days_since_month_end)
periods_df['days_till_month_end'] = periods_df['date'].apply(days_till_month_end)

periods_df['period'] = periods_df.apply(get_periods, axis=1) 
```

```python
data_df['period_event'] = data_df['period'].replace(2,1).replace(3,0).replace(4,0)
# new events
data_df['event_number'] = (data_df['period_event'].diff().replace(-1,0).cumsum()*data_df['period_event']).replace(np.nan,0).apply(int)
```

```python
def calc_drawdown_va(cva):
        #'''
        #cva = 1 + cva
        ## calculate cummax (max of 1 and cva)
        #maxcva = cva.cummax(axis=0)
        #maxcva[maxcva < 1] = 1
        ## calculate drawdown
        #drawdown = cva / maxcva
        ## center it around 0
        #drawdown = drawdown - 1
        #'''
        # calculate cummax
        maxcva = cva.cummax(axis=0)
        # calculate drawdown
        drawdown = cva - maxcva
        return drawdown

def get_returns_drawdowns(data_df):
    output_df = {} # returns
    intraday_drawdown_df = {}
    for grp_num, data in data_df.groupby('event_number'):
        if grp_num == 0:
            continue
        event_start = data.index.min()
        event_returns = data[tickers].sum() # apply(lambda x : log(1+x)).sum()
        output_df[event_start] = event_returns
        intraday_drawdown_df[event_start] = calc_drawdown_va(data[tickers].cumsum()).min() # apply(lambda x : log(1+x)).cumsum()).min()
    output_df = pd.DataFrame(output_df).T
    intraday_drawdown_df = pd.DataFrame(intraday_drawdown_df).T
    return output_df, intraday_drawdown_df
```

```python
tickers = ['SPY'] # 'BYD','PNK','SPY']
data_returns_df = data_df[tickers].dropna() # * multiplier
data_returns_df['event_number'] = data_df['event_number']

output_df, intraday_drawdown_df = get_returns_drawdowns(data_returns_df) # data_df
```
